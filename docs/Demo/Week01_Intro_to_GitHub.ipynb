### 2. Set up account name and email address

Set up your account name and email address in Git

The account name can be anyname, Git doesn't care. it just need to know to whom to credit commits and future projects. 

```bash
git config --global user.name "Double"
```

- More information pelease find here [Setting your username in Git](https://docs.github.com/en/get-started/getting-started-with-git/setting-your-username-in-git)

Use the sign-up email of your GitHub Account

```bash
git config --global user.email "GitHub_Sign_Up_EmailL"
```

### 5. Troubleshooting

5.1 Token issue

Error message:
_Remote: invalid username or password_
_fatal: Authentication failed_

Solution:

1. click your profile photo, then click Settings.
2. Navigate to Developer settings > Personal access tokens > Tokens (classic) > Generate new token > Generate new token classic
3. In the _Note_ field, assign a name to this token
4. In the _Expiration_ field, set an expiration date for this token. In this example, I set it for 90 days
5. From the __scopes__ menu, select all options. This will allow you to edit your repo



6. Click Generate token at the bottom of the page
7. Copy the generated token as it won't be shown again and temporatily same the token before closing the window



8. Changing a remote repository's URL

Replace <your-token> with the token you just copied

Replace USERNAME with your GitHub username

Replace REPO with the name of your GitHub Repo

```bash
git remote set-url origin git remote set-url origin https://<your-token>@github.com/USERNAME/REPO.git
```
